```python
import numpy as np
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../data/raw/loan_data.csv')
```

```python
df.head()
```

```python
df.drop("Loan_ID" , axis=1 , inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Married'].fillna(df["Married"].mode()[0], inplace=True)
df['Dependents'].fillna(df["Dependents"].mode()[0], inplace=True)
df["Self_Employed"].fillna(df["Self_Employed"].mode()[0], inplace=True)
```

```python
df['Credit_History'].fillna(df["Credit_History"].mode()[0], inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0], inplace=True)
```

```python
df['LoanAmount'].fillna(df['LoanAmount'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Dependents"] = df["Dependents"].replace("3+",3)
df["Dependents"] = df["Dependents"].astype(int)
```

```python
df.info()
```

```python
df["Gender"] = df["Gender"].map({"Male":1, "Female":0})
df["Married"] = df["Married"].map({"Yes":1 , "No":0})
df["Education"] = df["Education"].map({"Graduate":1 , "Not Graduate":0})
df["Self_Employed"] = df["Self_Employed"].map({"Yes":1 , "No":0})
df['Loan_Status'] = df["Loan_Status"].map({'Y':1 , 'N':0})
```

```python
df = pd.get_dummies(df, columns=['Property_Area'], drop_first=True)
```

```python
df = df.astype(int)
```

```python
df.head()
```

```python
df.to_csv("../data/processed/clean_data.csv", index=False)
```

# Data Preprocessing Summary

- Dropped unnecessary column (Loan_ID)
- Handled missing values using mode and median
- Fixed inconsistent values in Dependents
- Converted categorical data into numeric format
- Applied one-hot encoding for Property_Area
- Saved clean dataset for model training
